```python
import requests
import time
from pandas import json_normalize
```

```python
def fetch_page(api_url, page_number):
    params = {
        'from': '2024-06-23',
        'to': '2024-07-23',
        'q': '2024',
        'apiKey': 'fbe4cf3035f148c2a32bc54e5e5a9836',
        'pageSize': 100,  # Adjust based on the maximum allowed by the API
        'page': page_number
    }
    response = requests.get(api_url, params=params)
    time.sleep(1)  # Sleep 1 second between requests to manage rate limiting
    if response.status_code != 200:
        print(f"Error fetching page {page_number}: {response.status_code}")
        print(response.text)  # Print error response text
        return {}  # Return an empty dict to handle gracefully in the calling code
    return response.json()
```

```python
initial_data = fetch_page(api_url, 1)
if 'articles' in initial_data:
    total_results = initial_data['totalResults']
    articles = initial_data['articles']

    # Calculate the number of pages
    total_pages = (total_results + 99) // 100  # Assuming pageSize is 100

    # Fetch remaining pages
    for page in range(2, total_pages + 1):
        data = fetch_page(api_url, page)
        if 'articles' in data:
            articles.extend(data['articles'])
        else:
            print(f"No articles found on page {page}")

    # Normalize the collected articles into a DataFrame
    articles_df = json_normalize(articles)
    print(articles_df)
    print(articles_df.columns)
else:
    print("No articles found in initial request")
```
